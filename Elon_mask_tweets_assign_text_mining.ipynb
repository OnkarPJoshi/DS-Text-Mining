```python
import string
import re
def clean_text(text):
    text = text.lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\w*\d\w*', '', text)
    text = re.sub("[0-9" "]+"," ",text)
    text = re.sub('[]', '', text)
    
    return text

clean = lambda x: clean_text(x)
```

```python
df.Text =df.Text.apply(clean)
df.head(10)
```

```python
def remove_stopwords(text):
    no_stopwords = []
    for i in text.split():
        if i not in stop :
          if i not in stopwords:
            no_stopwords.append(i)
            
    x = ' '.join(no_stopwords)
    return x
```

```python
afinn = pd.read_csv('Afinn.csv', error_bad_lines=False, sep = ",", encoding = 'latin-1')
afinn.head(10)
```

```python
affinity_score = afinn.set_index('word')['value'].to_dict()
affinity_score
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
def calculate_sentiment(text):
  sent_score = 0
  if text:
    sentence = nlp(text)
    for i in sentence:
      sent_score += affinity_score.get(i.lemma_,0)

    return sent_score 
```

```python
calculate_sentiment('you are beautiful')
```

```python
df['sentiment_score'] = df['Text'].apply(calculate_sentiment)
df.head()
```

```python
df['word_count'] = df['Text'].str.split().apply(len)
df.head(10)
```
